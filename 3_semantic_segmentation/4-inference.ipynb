```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import torch
```

```python
from dataloader import makeDatapathList, dataTransform

root_path='./data/VOCdevkit/VOC2012/'

datapath_list=makeDatapathList(root_path)
train_img_list,train_anno_list=datapath_list('train')
val_img_list,val_anno_list=datapath_list('val')
```

```python
from pspnet import PSPNet

net=PSPNet(n_classes=21)
state_dict=torch.load('./weights/pspnet50_30.pth',map_location={'cuda:0':'cpu'})
net.load_state_dict(state_dict)
```

```python
img_file_path='./data/cowboy-757575_640.jpg'
img=Image.open(img_file_path)
img_width,img_height=img.size
plt.imshow(img)
plt.show()
```

```python
color_mean=(0.485,0.456,0.406)
color_std=(0.229,0.224,0.225)
transform=dataTransform(475,color_mean,color_std)
anno_file_path=val_anno_list[0]
anno_class_img=Image.open(anno_file_path)
p_palette=anno_class_img.getpalette()
img,anno_class_img=transform("val",img,anno_class_img)
```

```python
net.eval()
x=img.unsqueeze(0)
outputs=net(x)
y=outputs[0]
```

```python
y=y[0].detach().numpy()
y=np.argmax(y,axis=0)
anno_class_img=Image.fromarray(np.uint8(y),mode="P")
anno_class_img=anno_class_img.resize((img_width,img_height),Image.NEAREST)
anno_class_img.putpalette (p_palette)
plt.imshow(anno_class_img)
plt.show()
```

```python
trans_img=Image.new("RGBA",anno_class_img.size,(0,0,0,0))
anno_class_img=anno_class_img.convert("RGBA")
for x in range(img_width):
    for y in range(img_height):
        r,g,b,a=anno_class_img.getpixel((x,y))
        if (r,g,b)==(0,0,0):
            continue
        trans_img.putpixel((x,y),(r,g,b,150))
img=Image.open(img_file_path)
result=Image.alpha_composite(img.convert("RGBA"),trans_img)
plt.imshow(result)
plt.show()
```
